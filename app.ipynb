```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
es = pd.read_csv('ES.txt')
es["Datetime"] = pd.to_datetime(es["Date"] + es[" Time"])
es['range'] = abs(es[' High'] - es[' Low'])
es_adj = es.set_index("Datetime")
es_adj
```

```python
rangeDistribution = px.histogram(es,x="range",nbins=100,title="1-min Bar Range Frequency (ES)",marginal = 'box')

mean = np.mean(es['range'])
stdev_pluss = np.std(es['range'])
stdev_minus = np.std(es['range'])*-1

rangeDistribution.show()
```

```python
overnight = es_adj.between_time('18:00','9:29')
day = es_adj.between_time('9:30','16:00')
overnightHigh = np.max(overnight[" High"])
dayHigh = np.max(day[" High"])

if overnightHigh > dayHigh:
    print("Overnight hight not breached!")
else:
    print("Overnight high breached!" + str(overnightHigh) + ' ' + str(dayHigh))
```
